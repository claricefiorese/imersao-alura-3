### Desafio 01: Investigar por que a classe tratamento é tão desbalanceada



````python
dados["tratamento"].value_counts()
````

````python
#Filtrando o tratamento "com controle"
controle = dados.query('tratamento == "com_controle"')
controle["droga"].value_counts()
````

Em experimentos, só há necessidade de um grupo controle, independente da quantidade de tratamentos/drogas a serem testadas. E em todos os grupos de controle o mesmo tipo de droga é utilizado. Sendo assim, terá muito mais amostras com droga do que com controle.

### Desafio 03: Proporção das classes tratamento



````python
#Analisando a frequência em porcentagem
respostas = {
    "com_droga": "Proporção com droga",
    "com_controle": "Proporção com controle",
}

frequencia = dados["tratamento"].value_counts()

percentual = dados["tratamento"].value_counts(normalize = True) * 100

dist_freq = pd.DataFrame({"Frequência": frequencia, "Percentual": percentual })
dist_freq.rename(index = respostas, inplace = True)
dist_freq.rename_axis("Proporção das classes tratamento", axis=1, inplace = True) #coluna
dist_freq
````
